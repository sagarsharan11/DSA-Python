# Recursion

A function that calls itself directly or indirectly is called recursion. The corresponding function is called a recursive function

* Iteration / Loops ~ Recursion

Using recursion in these cases is effectiv rather that using iteration/loops ::
> * Recursion
> * Trees
> * Graphs

Basic Assumptions:
* Outer function perform some steps
* Inner function performs the rest of work

### Easy Questions on Recursion

```python
### Question Print Number from 5 to 1


print("Using Iteration / Loop")
for i in range(5,0,-1):
    print(i)

print("Using Recursion")
def printNos(n):
    if n==0:
        return
    print(n)
    printNos(n-1)
print(printNos(5))
```

```python
### Question: Print Factorial of number n

def factorial(n):
    if (n == 1 or n == 0):
        return 1
    return n * factorial(n-1) 
print(factorial(5))
```

```python
### Question: Print nth term of Fibonacci Sequence

def fibo(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    sum_nm2 = fibo(n-2)
    sum_nm1 = fibo(n-1)
    sum_n = sum_nm1 + sum_nm2
    return sum_n
print(fibo(6))
```

```python
### Question: Print x^n (stack height = n)

def stackHeight(x, n):
    if x == 0:
        return 0
    elif n == 0:
        return 1
    return x * stackHeight(x, n-1)
print(stackHeight(2, 4))
## Space Complexity - O(n)
```

```python
### Question: Print x^n (stack height = logn)

def stackHeight(x, n):
    if x == 0:
        return 0
    elif n == 0:
        return 1
    ## n-even
    if n % 2==0:
        return stackHeight(x, n//2) * stackHeight(x, n//2)
    ## n-odd
    else:
        return stackHeight(x, n//2) * stackHeight(x, n//2) * x

print(stackHeight(2, 4))
## Space Complexity - logn | using Divide and Conquer method
```

```python
### Question: Tower of Hanoi | 3 towers are there
## Rule1 - Only one disk transferred in 1 step
## Rule2 - Smaller disks are always kept on topof laeger disks



def towerOfHanoi(n, src, helper, dest):
    if n==1:
        print("transfer disk ",n, " from ", src, " to ", dest)
        return
    towerOfHanoi(n-1, src, dest, helper)
    print("transfer disk ",n, " from ", src, " to ", dest)
    towerOfHanoi(n-1, helper, src, dest)
print(towerOfHanoi(3, "S", "H", "D"))
```

```python
### Question: Find the last and first occurence of an element in a string

class Occurences:
    def __init__(self):
        self.first = -1
        self.last = -1

    def FirstLastString(self, idx, s, element):
        if idx == len(s):
            print(self.first)
            print(self.last)
            return
        currChar = s[idx]
        if currChar == element:
            if self.first == -1:
                self.first = idx
            else:
                self.last = idx
        self.FirstLastString(idx+1, s, element)
```

```python
### Question: Check if an array / list is sorted (Strictly Increasing)

def sortedStricltyInc(idx, arr):
    if idx == len(arr)-1:
        return True
    if arr[idx] < arr[idx+1]:
        return sortedStricltyInc(idx + 1, arr)
    else:
        return False

print(sortedStricltyInc(0, [1,2,3,4,5]))
```

```python
### Question: Move all 'x' to the end of the string

def moveEnd(idx, counter, s):
    if idx == len(s):
        print('x'*counter)
        return
    if s[idx] == 'x':
        counter += 1
        moveEnd(idx+1, counter, s)
    elif s[idx] != 'x':
        print(s[idx], end="")
        moveEnd(idx+1, counter, s)
    
print(moveEnd(0, 0, 'axbcxxd'))
```

```python
def removeDuplicates(idx, s, newStr):
    if idx == len(s):
        print(newStr)
        return
    if s[idx] not in newStr:
        newStr += s[idx]
        removeDuplicates(idx + 1, s, newStr)
    else:
        removeDuplicates(idx + 1, s, newStr)

print(removeDuplicates(0, 'abbccda', ''))
```

```python
### Question: Print all the subsequence of a string
# str = 'abc' | subsequence = a, b, c, ab, ac, bc, abc

def subsequences(s, idx, newStr):
    if idx == len(s):
        print(newStr)
        return
    subsequences(s, idx+1, newStr+s[idx])
    subsequences(s, idx+1, newStr)
print(subsequences('abc', 0, ""))
```

```python
### Question: Print all the unique subsequence of a string
# str = 'aaa' | subsequence = aaa, aa, aa, a, aa, a, a, _

def subsequences(s, idx, newStr, str_set):
    if idx == len(s):
        if newStr in str_set:
            return
        else:
            print(newStr)
            str_set.add(newStr)
            return
    subsequences(s, idx+1, newStr+s[idx], str_set)
    subsequences(s, idx+1, newStr, str_set)
str_set = set()
print(subsequences('aaa', 0, "", str_set))
```

```python
### Question: Print keypad combination

keypad = ['.','abc','def','ghi','jkl','mno','pqrs','tu','vwx', 'yz']
def printComb(s, idx, combination):
    if idx == len(s):
        print(combination)
        return
    mapping = keypad[int(s[idx])]
    for i in mapping:
        printComb(s, idx+1, combination + i)

print(printComb('23', 0, ''))
```

```python
### Print all permutations of a string

def printPermutations(s, permutation):
    if len(s) == 0:
        print(permutation)
        return
    for i in range(0,len(s)):
        newStr = s[0:i] + s[i+1:]
        printPermutations(newStr, permutation + s[i])
print(printPermutations('abc', ''))

### Time Complexity O(n!)
```

```python
### Count total paths in a maze to move from (0,0) to (n,m), n=3, m=3
### Condition is move only in right and down

def countPaths(i,j,n,m):
    if (i == n) or (j == m):
        return 0

    if (i == n-1) & (j== m-1):
        return 1

    ## move downwards
    downPath = countPaths(i+1, j, n, m)

    ## move right
    rightPath = countPaths(i, j+1, n, m)

    return downPath + rightPath


print(countPaths(i=0,j=0,n=3,m=3))
```

```python
### Place Tiles of size 1 x m in a floor of size n x m, n=4, m=2
### Tiles can be placed horizontally or vertically..

def placeTiles(n,m):
    if n == m:
        return 2
    if n < m:
        return 1

    ##vertical
    vertPlace = placeTiles(n-m,m)

    ##horizontal
    horzPlace = placeTiles(n-1,m)

    return vertPlace + horzPlace

print(placeTiles(n=4,m=2))
```

```python
def inviteGuests(n):
    if n <=1:
        return 1

    indi = inviteGuests(n-1)
    pairs = (n-1) * inviteGuests(n-2)

    return indi + pairs

print(inviteGuests(n=4))
```

# Backtracking

Find all possible solutions and use the one you want, i.e., Pehle istemaal krein fir vishwas krein

```python
def printPermutation(s, perm, idx):
    if len(s) == 0:
        print(perm)
        return

    for i in range(len(s)):
        currChar = s[i]
        new_s = s[0:i] + s[i+1:]
        printPermutation(new_s, perm + currChar, idx+1)
```
